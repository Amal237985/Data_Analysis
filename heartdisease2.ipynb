```python
# importing libraries that is being used and read the file and save in dataframe
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import plotly.express as px
df=pd.read_csv('/content/HeartDisease.csv')
df
```

```python
# Removing unwanted data and fill 0 for NAN values
df0=df.drop(axis=1,columns=['Year','LocationAbbr','DataSource','Data_Value_Footnote_Symbol','Data_Value_Footnote','TopicID','LocationID'])
df1=df0.fillna(0)
df1
```

```python
# county with most heart diseases.Analysing by checking the overall death rate of counties for overall gender and race.
b0=df1[df1['GeographicLevel']=='County']
b00=b0[b0['Stratification1']=='Overall']
b01=b00[b00['Stratification2']=='Overall']
b02=b01[b01['Data_Value']==b01['Data_Value'].max()][['LocationDesc','Data_Value','Stratification1','Stratification2']]
print(b02)
b03=b01.sort_values(by='Data_Value',ascending=False)
b03.head(10)
```

***By analyzing the heart disease mortality rate of counties for overall gender and race/ethnicity we list out the top 20 states with high mortality.More inspections to be done in these couties regarding life styles,living conditions etc.to study on the reasons regarding for these higher mortality rate and to find effective methods to reduce those.***

```python
# Gender-Race  highest 20 Deathrated Counties(for each pair of gender-race pair)
c0=df2[df2['GeographicLevel']=='County']
strat1=c0['Stratification1'].unique().tolist()#[male,female,overall]
strat2=c0['Stratification2'].unique().tolist()#[black,white,.....]
for i in strat1:
  for j in strat2:
    c00=c0[c0['Stratification1']==i]
    c01=c00[c00['Stratification2']==j].sort_values(by=['Data_Value'],ascending=False).head(10)
    x=c01['LocationDesc'].tolist()
    y=c01['Data_Value'].tolist()
    z=pd.DataFrame({'deathrate':y},index=x)
    print('highest '+i+"-"+j+" death rate\n",z,"\n\n")
    z.plot.bar(title=j+"-"+i,xlabel='County',ylabel='Death rate',color='grey')
```

***Here we analyzed the data based on categorizing each gender and race\ethnicity pair and come up with the highest death rated counties with each  gender and race\ethnicity pair.
```python
# Gender-Race  highest 10 Deathrated States(for each pair of gender-race pair)
e0=df2[df2['GeographicLevel']=='State']
strat1=e0['Stratification1'].unique().tolist()#[male,female,overall]
strat2=e0['Stratification2'].unique().tolist()#[black,white,.....]
for i in strat1:
  for j in strat2:
    e00=e0[e0['Stratification1']==i]
    e01=e00[e00['Stratification2']==j].sort_values(by=['Data_Value'],ascending=False).head(10)
    x=e01['LocationDesc'].tolist()
    y=e01['Data_Value'].tolist()
    z=pd.DataFrame({'deathrate':y},index=x)
    print('highest '+i+"-"+j+" death rate\n",z,"\n\n")
    z.plot.bar(title='highest '+j+"-"+i+' death rate')
```

```python
# 1) US nation deathrate of different race compared with gender.
gen=['Male','Overall','Female']
strat2=['Black','White','Hispanic','Asian and Pacific Islander','American Indian and Alaskan Native','Overall']
f=df1[df1['GeographicLevel']=='Nation']
c=[]
for i in gen:
  b={}
  for j in strat2:
    f1=f[f['Stratification1']==i]
    b[j]=f1[f1['Stratification2']==j]['Data_Value'].sum()
  c.append(b)
df00=pd.DataFrame({'Black':[c[0]['Black'],c[1]['Black'],c[2]['Black']],'White':[c[0]['White'],c[1]['White'],c[2]['White']],'Hispanic':[c[0]['Hispanic'],c[1]['Hispanic'],c[2]['Hispanic']],'Asian and Pacific Islander':[c[0]['Asian and Pacific Islander'],c[1]['Asian and Pacific Islander'],c[2]['Asian and Pacific Islander']],'American Indian and Alaskan Native':[c[0]['American Indian and Alaskan Native'],c[1]['American Indian and Alaskan Native'],c[2]['American Indian and Alaskan Native']],'Overall':[c[0]['Overall'],c[1]['Overall'],c[2]['Overall']]},index=gen)
df00.plot.bar(rot=0,xlabel='Gender',ylabel='Death Rate',title='Death rate per 100000 in US')
```

```python
# 2)
# For checking about the geographic location of counties of most heart diseases to analyze is there anything common.
#Converting the location column(given string) into geomerty column for using geopandas
df11=df1[df1['GeographicLevel']=='County']
lat=[]
long=[]
locat=df11['Location 1'].tolist()
for i in range(len(locat)):
    if locat[i]==0:
      lat.append(0)
      long.append(0)
    else:
      z=float(locat[i].split(',')[0].split('(')[1])
      lat.append(z)
      w=float(locat[i].split(',')[1].split(')')[0])
      long.append(w)
df11['latitude']=lat
df11['longitude']=long
df11
```
