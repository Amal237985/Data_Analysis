```python
# 18. Are there any correlations between different attributes and the likelihood of subscribing to a term deposit?
numerical_y=[]
for i in df['y'].tolist():
  if i=='no':
    numerical_y.append(0)
  elif i=='yes':
    numerical_y.append(1)
df['numerical_y']=numerical_y
df.head(10)
```

```python
df.corr(numeric_only=True)['numerical_y'].sort_values()
```

```python
from scipy.stats import chi2_contingency
attributes=df.columns
correlated=[]
for i in attributes:
  ctr=pd.crosstab(index=df[i],columns=df['y'])
  Chisqresult=chi2_contingency(ctr)
  if Chisqresult[1]<0.05:
    correlated.append(i)
print(correlated)
```

Since we have to find the correlations between columns with categorical values, here i use Chi sqaure test .
In which we take, null hypothesis H0 : Columns are not correlated.
Now finding the p-value and if it is less than 0.05 (95% confidence ) we reject H0 and there by there exist a correlation between the columns ad if p-value is greater than 0.05  accept H0 there by there doesnot exist a correlation between the columns.
Using this method here i get all the attributes are correlated with the target column y.
