```r
# Win percentage of Teams
teams<-list(unique(df$BattingTeam))
E<-list()
G<-list()
for (j in teams[[1]]){
E<-append(E,j)
df1<-subset(df,BattingTeam==j)
b<-list(unique(df1$ID))
z<-0
  for (i in 0:length(b[[1]])+1){
  x1<-0
  x2<-0
  df2<-subset(df,ID==b[[1]][i])
  df3<-subset(df2,innings==1)
  x1<-x1+(sum(df3$total_run))
  df4<-subset(df2,innings==2)
  x2<-x2+sum(df4$total_run)
  if (x1>x2){
    if (list(df3$BattingTeam)[[1]][1]==j){
      z<-z+1
    }
  }
  if (x2>x1) {
     if (list(df4$BattingTeam)[[1]][1]==j){
      z<-z+1
    }
  }
}
win<-(z/length(b[[1]]))*100
G<-append(G,win)
}

TEAM<-unlist(E)
WIN_PERCENTAGE<-unlist(G)
DF<-data.frame(TEAM,WIN_PERCENTAGE)
DF
```

```r
#consistend bowler
bowlers<-list(unique(df$bowler))
r_conc <-list()
wick_tak <-list()
balls<-list()
for (i in bowlers[[1]]){
  df7=subset(df,bowler==i)
  r_conc<-append(r_conc,sum(df7$total_run))
  balls<-append(balls,length(df7$ballnumber))
  df8=subset(df7,isWicketDelivery==1)
  df9=subset(df7,kind!="run out")
  wick_tak<-append(wick_tak,length(df9$isWicketDelivery))
}
bowl_avg<-list()
overs<-list()
eco<-list()
for (j in 1:length(bowlers[[1]])){
  ss<-(r_conc[[j]])/(wick_tak[[j]])
  bowl_avg<-append(bowl_avg,ss)
  overs<-append(overs,(balls[[j]]/6))
}
for (k in 1:length(bowlers[[1]])){
  SSS<-(r_conc[[k]])/(overs[[k]])
  eco<-append(eco,SSS)
}
RUNS=unlist(r_conc)
WICKETS=unlist(wick_tak)
BOWLERS=unlist(bowlers)
AVERAGE=unlist(bowl_avg)
OVERS=unlist(overs)
ECONOMY=unlist(eco)
DF11<-data.frame(BOWLERS,RUNS,WICKETS,OVERS,AVERAGE,ECONOMY)
DF11
```

```r
#Extra balls delivered bowler
b1<-list(unique(df$bowler))
count_extra<-list()
for (i in b1[[1]]){
  extraball<-0
  c1<-subset(df,bowler==i)
  b2<-list(unique(c1$ID))
  for (j in b2[[1]]){
    c2<-subset(c1,ID==j)
    b3<-list(unique(c2$overs))
    for (k in b3[[1]]){
      c3<-subset(c2,overs==k)
      extraball=extraball+(length(c3$ballnumber)-6)
    }
  }
 count_extra<-append(count_extra,extraball)
}
b11<-unlist(b1)
extra_balls<-unlist(count_extra)
DF00<-data.frame(b11,extra_balls)
DF000<-arrange(DF00,-extra_balls)
head(DF000,25)
```

```r
# Define the batter and bowler
batter <- "JC Buttler"
bowler <- "Yash Dayal"

# Filter the data frame to get the relevant rows
batter_bowler_data <- df[df$batter == batter & df$bowler == bowler, ]

# Calculate the runs made, balls, and wickets
runs_made <- sum(batter_bowler_data$total_run)
balls <- nrow(batter_bowler_data)
wickets <- nrow(batter_bowler_data[batter_bowler_data$isWicketDelivery == 1, ])

# Calculate the economy rate
economy <- (runs_made / balls) * 6

# Calculate the batting average
if (wickets != 0) {
  batting_avg <- runs_made / wickets
} else {
  batting_avg <- "Infinity"
}

# Create a data frame to store the stats
stats <- data.frame(
  "Stats for" = paste(batter, "vs", bowler),
  "Balls" = balls,
  "Runs Scored" = runs_made,
  "Economy" = economy,
  "Batting Avg" = batting_avg
)

# Print the stats
print(stats)
```

```r
# highest individual score
player=list(unique(df$batter))
h_score=0
for (i in player[[1]])
{
  a<-df[df$batter==i,]
  id<-list(unique(a$ID))
  for (j in id[[1]])
  {
    match<-a[a$ID==j,]
    if (sum(match$batsman_run)>h_score){
      h_score=sum(match$batsman_run)
    }
  }
}
print(h_score)
```
